```python
# fill Employee_Name with EmpID as string \\bc there is 70% have a names and we not want to drop a column,
#and i use this column in other case.
df['Employee_Name'] = df['Employee_Name'].fillna(df['EmpID'])
```

```python
df['Employee_Name'] = df['Employee_Name'].astype(str)
```

```python
df['Employee_Name'].info()
```

```python
# fill salary with same salary avrage of same Position
# secound step we take same Department
# therd step we take mean for salary
#\\bc we if we use mean at all salary we gana have biased 
```

```python
#now we want to know each avrage postion salary:
salary_avg_by_P = df.groupby('Position')['Salary'].transform('mean')
df['Salary'] = df['Salary'].fillna(salary_avg_by_P)
```

```python
salary_avg_by_P.sort_values()
```

```python
df['Salary'].info()
```

```python
salary_avg_by_D= df.groupby('Department')['Salary'].transform('mean')
df['Salary'] = df['Salary'].fillna(salary_avg_by_D)
```

```python
df['Salary'].info()
```

```python
df[df['Salary'].isna()]
```

```python
df['Salary'] = df['Salary'].fillna(df['Salary'].mean())
```

```python
df.info()
```

```python
# fill DOB with Unknown   \\ bc if we fill it with mean we will have baies 
df['DOB'] = df['DOB'].fillna('Unknown')
```

```python
df.info()
```

```python
# fill sex with same sex of first name of Employee_Name 
#\\ and if we not find a same name , then we take a mean sex of each Department, i try it but not fond any duplecate first name 
# now i try to take a sex mean of position groupby 
# idont know how to write a code (df.groupby('Position').groupby('Sex'))
#then i will try to fill it with Unknown

df['Sex'] = df['Sex'].fillna('Unknown')
```

```python
df.info()
```

```python
# drop DateofTermination \\bc there is 66% of column is missing
df = df.drop('DateofTermination',axis=1)
```

```python
df.info()
```

```python
# fill ManagerID with bfill, bc there is a litl set of names and no uniqe numbers of ids in this data set..
mm = df[df['ManagerID'].isna()]
mm
```

```python
df['ManagerID'] = df['ManagerID'].fillna(method='bfill')
```

```python
df.info()
```

```python
# fill Absences with mean of same Department \\bc i think pepole teach each other
avg_of_Department_Absences= df.groupby('Department')['Absences'].transform('mean')
df['Absences'] = df['Absences'].fillna(avg_of_Department_Absences)
```
