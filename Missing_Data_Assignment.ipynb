```python
df.isnull().sum()
```

```python
df.drop(['Employee_Name'] ,axis = 1, inplace=True)
```

```python
#since we have the means of postions we can fill based on postion, but for now i will just fill with mean
df['Salary'].fillna(df['Salary'].mean(), inplace=True)
```

```python
from datetime import datetime
df['DOB'] = pd.to_datetime(df['DOB'], format='%m/%d/%y')
today = datetime.now()
def age(dob):
    return today.year - dob.year - ((today.month, today.day) < (dob.month, dob.day))
df['Age'] = df['DOB'].apply(age)
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.drop(['DOB'] ,axis = 1, inplace=True)
```

```python
# I have calculated the percentage of males and females of the dataset and filled them based on that
df['Sex'].fillna('F',limit=36,  inplace=True)
df['Sex'].fillna('M',limit=26,  inplace=True)
```

```python
df.drop(['DateofTermination'] ,axis = 1, inplace=True)
```

```python
df.drop(['ManagerID'] ,axis = 1, inplace=True)
```

```python
df['Absences'].fillna(df['Absences'].mode()[0], inplace=True)
```
